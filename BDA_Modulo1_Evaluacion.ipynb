```python
inventario = [{"nombre": "botas", "precio": 30, "cantidad": 100}, {"nombre" : "chanclas", "precio": 10, "cantidad": 200}, {"nombre" : "chaleco", "precio" : 45, "cantidad" : 10}]
clientes = {"nombre" : "Mireia Pardo", "email" : "mireia.p.d@hotmail.com", "compras" : []}, {"nombre" : "Lia Kinkladze", "email" : "lia1997@hotmail.com", "compras" : []}, {"nombre" : "Ramón Romero", "email" : "ramon87@hotmail.es", "compras" : []}
ventas_totales = 0
```

```python
def agregar_producto(nombre, precio, cantidad):
    for producto in inventario:
        if producto["nombre"] == nombre:
            producto["cantidad"] += cantidad
            return

    inventario.append({"nombre": nombre, "precio": precio, "cantidad": cantidad})
    
```

```python
agregar_producto("botas",30,10)
inventario
```

```python
agregar_producto("chaqueta",25,10)
```

```python
def ver_inventario():
    for producto in inventario:
        print(f"Nombre: $ {producto['nombre']},"
            f"Precio: ${producto['precio']},"
            f"Cantidad:${producto['cantidad']}")
    
```

```python
ver_inventario()
```

```python
def buscar_producto(nombre):
    for producto in inventario:
        if producto ["nombre"] == nombre:
            print(
                f"Nombre: {producto['nombre']}, "
                f"Precio: ${producto['precio']}, "
                f"Cantidad: {producto['cantidad']}")
            return
```

```python
buscar_producto("botas")
```

```python
def actualizar_stock(nombre,cantidad):
    for producto in inventario:
        if producto["nombre"] == nombre:
            producto["cantidad"] += cantidad
            print(
                f"Stock actualizado → "
                f"Nombre: {producto['nombre']}, "
                f"Cantidad actual: {producto['cantidad']}"
            )
            return

    print("El producto no está en el inventario")
```

```python
actualizar_stock("botas",10)
```

```python
def eliminar_producto(nombre):
    for producto in inventario:
        if producto["nombre"] == nombre:
            inventario.remove(producto)
            print(f"Producto '{nombre}' eliminado del inventario")
            return

    print("El producto no está en el inventario")
```

```python
eliminar_producto("botas")
```

```python
eliminar_producto("camiseta")
```

```python
def calcular_valor_inventario():
    valor_total = 0
    for producto in inventario:
        valor_producto = producto["precio"] * producto["cantidad"]
        valor_total += valor_producto

    print(f"El valor total del inventario es: {valor_total} euros")

```

```python
calcular_valor_inventario()
```
